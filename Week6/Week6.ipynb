```python
# IMPORT STATEMENTS
# 1. Import math

import math

# FUNCTION DEFINITIONS
# 1. Define area(radius): returns pi * r squared

def circle_area(radius):
  area = math.pi * (radius ** 2)
  return area

# 2. Define circumference(radius): returns 2 * pi * r

def circle_circumference(radius):
  circumference = 2 * math.pi * radius
  return circumference

# 3. Define volume(radius, height):

def cylinder_volume(radius, height):
  volume = math.pi * (radius ** 2) * height
  return volume


# PROGRAM LOGIC
# 1. Input for radius (float)

input_radius = float(input("Enter radius:"))

# 2. Call and print area and circumference.

area = circle_area(input_radius)
print("Area:", area)

circumference = circle_circumference(input_radius)
print("Circumference:", circumference)


# 3. Input for height (float)

input_height = float(input("Enter height:"))

# 4. Call and print volume.

volume = cylinder_volume(input_radius, input_height)
print("Volume:", volume)
```
