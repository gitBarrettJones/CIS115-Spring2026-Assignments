```python
# NAME: rentalcar.py
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS #

## Create Car Classifications with prices ##
# car_class (B, S, L) #
# check for case insensitivity #
# take the input and convert into uppercase .upper()#
def car_class():
  car = input("What class of car would you like? (B, S, L): ").upper()

  while car not in ["B", "S", "L"]:
    print("Invalid input. Please try again.")
    car = input("What class of car would you like? (B, S, L): ").upper()

  return car
# get_price #
def get_price(car):
# B , Budget = 40 #
  if car == "B":
    return 40, "Budget"
# S , Standard = 60 #
  elif car == "S":
    return 60, "Standard"
# L , Luxury = 80 #
  else:
    return 80, "Luxury"

## Obtain Mileage ##
# mileage(miles) #
def mileage(miles):
# First 100 are free, take miles and subtract by 100 #
  if miles > 100:
    return miles - 100
  else:
    return 0

## Total Cost ##
# Fruitful function to calculate the cost #

# PROGRAM LOGIC

## print the title and car classifications ##
print("Car Rental")
print("Enter B for Budget ($40 daily)")
print("Enter S for Standard ($60 daily)")
print("Enter L for Luxury ($80 daily)")
# gather user input for a valid car class#
car = car_class()

## Gather how many days the user will need the car ##
days = int(input("How many days would you like to rent the car? "))

## Get the price and classification name ##
price, classification = get_price(car)

## Get the mileage ##
# print the Mileage Policy #
print("Mileage Policy")
print("First 100 miles are free.")
print("Any additional miles will be $0.35 per mile.")

# Odometer Readings #
start_odometer = int(input("What is the starting odometer reading? "))
end_odometer = int(input("What is the ending odometer reading? "))

# Find the miles traveled #
miles = end_odometer - start_odometer

#Calculate mileage fee #
mileage_fee = mileage(miles)

## Total the final bill ##


## Display Results
print("Rental Summary")
print("Car Classification", classification)
print("Days Rented", days)
print("Mileage Fee", mileage_fee)
```
