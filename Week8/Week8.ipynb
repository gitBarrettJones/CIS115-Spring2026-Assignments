```python
#### IMPORT STATEMENTS

#### FUNCTION DEFINITIONS

# Define is_palindrome(word):
# Check for case insensitivity
#   Take the word and convert into lowercase using word.lower()
# Use word[::-1] negative index string operator to reverse the word
# Return True is word is a palindrome otherwise false.


#### PROGRAM LOGIC
# Take user input for a word
#   user_word = input()
# Call the function using user input
#   result = is_palindrome()
# Return value and display result


def is_palindrome(word):

    word = word.lower()
    if word == word[::-1]:
        return True
    else:
        return False

user_word = input("Enter your word please: ")

result = is_palindrome(user_word)

if result == True:
    print("Your word is a palindrome")
else:
    print("Your word is not a palindrome")
```
